```python
import pandas as pd
```

```python
metacritic_df = pd.read_csv("game_info_subset_adj.csv")
vgsales_df = pd.read_csv("vgsales.csv")
```

metacritic csv

```python
metacritic_df = metacritic_df.drop(["Unnamed: 0"], axis=1)
metacritic_df.to_csv('metacritic.csv', index=False)
metacritic_df.head(5)
```

genre sales csv

```python
vgsales_df.head(5)
```

```python
# vgsales_df = vgsales_df.dropna()
```

```python
# vgsales_df[['Genre','NA_Sales','EU_Sales','JP_Sales','Other_Sales']].to_csv('sales.csv', index=False)
genre_sales_df = vgsales_df[['Genre','NA_Sales','EU_Sales','JP_Sales','Other_Sales', 'Global_Sales']].groupby('Genre').agg("sum").reset_index()
genre_sales_df = genre_sales_df.sort_values('Global_Sales', ascending=False)
genre_sales_df.to_csv('genre_sales.csv', index=False)
genre_sales_df.head(5)
```

year_sales csv

```python
year_sales_df = vgsales_df[['Year','NA_Sales','EU_Sales','JP_Sales','Other_Sales', 'Global_Sales']].groupby('Year').agg("sum").reset_index()
year_sales_df = year_sales_df.sort_values('Year', ascending=False)
year_sales_df.to_csv('year_sales.csv', index=False)
year_sales_df.head(5)
```

publisher_sales csv

```python
publisher_sales_df = vgsales_df[['Publisher','NA_Sales','EU_Sales','JP_Sales','Other_Sales', 'Global_Sales']].groupby('Publisher').agg("sum").reset_index()
publisher_sales_df = publisher_sales_df.sort_values('Global_Sales', ascending=False)
publisher_sales_df.to_csv('publisher_sales.csv', index=False)
publisher_sales_df.head(5)
```

platform_sales csv

```python
platform_sales_df = vgsales_df[['Platform','NA_Sales','EU_Sales','JP_Sales','Other_Sales', 'Global_Sales']].groupby('Platform').agg("sum").reset_index()
platform_sales_df = platform_sales_df.sort_values('Global_Sales', ascending=False)
platform_sales_df.to_csv('platform_sales.csv', index=False)
platform_sales_df.head(5)
```

genre_ratings csv

```python
metacritic_df = vgsales_df.merge(metacritic_df, 'left', left_on='Name', right_on='name', )
metacritic_df = metacritic_df.dropna()
```

```python
metacritic_df = metacritic_df[['Genre','metacritic']].groupby('Genre').agg("mean").reset_index()
metacritic_df = metacritic_df.sort_values('metacritic', ascending=False)
metacritic_df.to_csv('genre_ratings.csv', index=False)
metacritic_df
```
