```python
import pandas as pd
```

```python
metacritic_df = pd.read_csv("game_info_subset_adj.csv")
vgsales_df = pd.read_csv("vgsales.csv")
```

metacritic csv

```python
metacritic_df = metacritic_df.drop(["Unnamed: 0"], axis=1)
metacritic_df.head(5)
```

```python
# vgsales_df[['Genre','NA_Sales','EU_Sales','JP_Sales','Other_Sales']].to_csv('sales.csv', index=False)
genre_sales_df = vgsales_df[['Genre','NA_Sales','EU_Sales','JP_Sales','Other_Sales', 'Global_Sales']].groupby('Genre').agg("sum").reset_index()
genre_sales_df = genre_sales_df.sort_values('Global_Sales', ascending=False)
genre_sales_df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales', 'Global_Sales']] = round(genre_sales_df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales', 'Global_Sales']], 2)
genre_sales_df.to_csv('genre_sales.csv', index=False)
genre_sales_df.head(5)
```

genre_ratings csv

```python
metacritic_df = vgsales_df.merge(metacritic_df, 'left', left_on='Name', right_on='name', )
metacritic_df = metacritic_df.dropna()
```

```python
mean = metacritic_df[['Genre','metacritic']].groupby('Genre').agg("mean").reset_index()
mean.rename(columns = {'metacritic':'meta_mean'}, inplace = True)
mean['meta_mean'] = round(mean['meta_mean'], 2)

max = metacritic_df[['Genre','metacritic']].groupby('Genre').agg("max").reset_index() #[['metacritic']]
max.rename(columns = {'metacritic':'meta_max'}, inplace = True)

min = metacritic_df[['Genre','metacritic']].groupby('Genre').agg("min").reset_index()#[['metacritic']]
min.rename(columns = {'metacritic':'meta_min'}, inplace = True)

median = metacritic_df[['Genre','metacritic']].groupby('Genre').agg("median").reset_index()#[['metacritic']]
median.rename(columns = {'metacritic':'meta_median'}, inplace = True)
median['meta_median'] = round(median['meta_median'], 2)

genre_metacritic_df = mean.merge(max, how='inner', on='Genre').merge(min, how='inner', on='Genre').merge(median, how='inner', on='Genre')
genre_metacritic_df = genre_metacritic_df.sort_values('meta_mean', ascending=False)

genre_metacritic_df.to_csv('genre_ratings.csv', index=False)
genre_metacritic_df 
```

meta_sales csv

```python
meta_sales_df = metacritic_df[['Global_Sales', 'metacritic']]
meta_sales_df[['Global_Sales']] = round(meta_sales_df[['Global_Sales']], 2)
meta_sales_df = meta_sales_df[meta_sales_df['Global_Sales'] < 80] # exclude single large outlier to clarify plot
meta_sales_df.to_csv('meta_sales.csv', index=False)
meta_sales_df.head()
```
